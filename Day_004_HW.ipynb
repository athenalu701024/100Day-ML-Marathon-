```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv') #將 dirpath 與 filename 結合，無需自行判斷要用 '\\' 或 '/' 
print('Path of read in data: %s' % (f_app)) #讀出f_app的路徑位置
app_train = pd.read_csv(f_app) #定義app_train為使用pd讀取csv
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
# .loc[:] 列出指定的行數
app_train.loc[1:10]
```

```python
# 查閱每一個變數彼此之間的相關係數
corr_app = app_train.corr()
corr_app
```

```python
# 相關係數圖表化，顏色越深表示相關信越強
plt.figure(figsize=(16,16))
sns.heatmap(corr_app, cmap='YlGnBu')
```
